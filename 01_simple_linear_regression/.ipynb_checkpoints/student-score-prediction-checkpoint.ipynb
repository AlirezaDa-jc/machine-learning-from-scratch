# Linear Regression Manual

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
X = np.array([1.5,3,4.5,6,7.5])  # feature: Study Hours
y = np.array([35,50,65,80,95])  # target: Score

# Add bias term (column of ones)
X_b = np.c_[np.ones((X.shape[0], 1)), X]

# Calculate weights (theta) using normal equation
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)

print("X_b", X_b)
print("Intercept (theta_0):", theta_best[0])
print("Slope (theta_1):", theta_best[1])

# Prediction
def predict_score(size):
    return theta_best[0] + theta_best[1] * size

new_hours = 4
predicted_price = predict_score(new_hours)
print(f"Predicted Score for Study Hours {new_hours}: {predicted_price:.2f} Score")

# Plot data points and regression line
plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X, theta_best[0] + theta_best[1] * X, color='red', label='Regression line')
plt.xlabel("Study Hours")
plt.ylabel("Score")
plt.legend()
plt.show()

# Predictions for all training data
y_pred = theta_best[0] + theta_best[1] * X

# Mean Squared Error
mse = np.mean((y - y_pred) ** 2)
print(f"Mean Squared Error (MSE): {mse:.2f}")

# Mean Absolute Error
mae = np.mean(np.abs(y - y_pred))
print(f"Mean Absolute Error (MAE): {mae:.2f}")
```

# Linear Regression Scikit

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Reshape X for sklearn (2D array)
X_reshaped = X.reshape(-1, 1)

# Train sklearn linear regression
model = LinearRegression()
model.fit(X_reshaped, y)

# Predict
y_pred_sklearn = model.predict(X_reshaped)

# Calculate errors
mse_sklearn = mean_squared_error(y, y_pred_sklearn)
mae_sklearn = mean_absolute_error(y, y_pred_sklearn)

print(f"Sklearn MSE: {mse_sklearn:.2f}")
print(f"Sklearn MAE: {mae_sklearn:.2f}")

print(f"Intercept (sklearn): {model.intercept_}")
print(f"Slope (sklearn): {model.coef_[0]}")

new_hours_array = np.array([[4]])
predicted_score_sklearn = model.predict(new_hours_array)
print(f"Sklearn predicted Score for Study Hours {new_hours}: {predicted_score_sklearn[0]:.2f} Score")
```
