# Student Performance Prediction - Notebook Script
It implements Multiple Linear Regression from scratch, compares with scikit-learn, and provides an interactive widget-based predictor using ipywidgets.

## 0. Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, Markdown

# For scikit-learn comparison
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# For interactive widgets
import ipywidgets as widgets
```

## 1. Load Dataset
The dataset columns expected: Hours_Studied, Sleep_Hours, Past_Score, Final_Exam_Score


```python
df = pd.read_csv('dataset.csv')

# Show basic info
print('Dataset loaded:')
display(df.head())
print('\nDataset shape:', df.shape)
```

## 2. Exploratory Data Analysis (EDA)

```python
print('\n--- Descriptive statistics ---')
display(df.describe())

plt.figure(figsize=(6,5))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# Pairplot (might take a moment)
try:
    sns.pairplot(df)
    plt.show()
except Exception as e:
    print('Pairplot failed (headless env?), continuing...')
```

## 3. Manual Multiple Linear Regression (Normal Equation)

```python
# Prepare X and y
feature_cols = df.columns[:-1].tolist()  # all except last
target_col = df.columns[-1]

X = df[feature_cols].values.astype(float)  # shape (m, n)
y = df[target_col].values.astype(float)    # shape (m,)

# Add bias column of ones
m = X.shape[0]
X_bias = np.c_[np.ones((m,1)), X]  # shape (m, n+1)

# Compute theta using Normal Equation: theta = (X^T X)^-1 X^T y
XT = X_bias.T
theta_manual = np.linalg.inv(XT.dot(X_bias)).dot(XT).dot(y)

print('\nManual coefficients (theta):')
for i, val in enumerate(theta_manual):
    if i == 0:
        print(f'  theta_0 (intercept): {val:.6f}')
    else:
        print(f'  theta_{i} ({feature_cols[i-1]}): {val:.6f}')

# Predictions using manual model
y_pred_manual = X_bias.dot(theta_manual)

# Evaluation (manual)
mse_manual = mean_squared_error(y, y_pred_manual)
r2_manual = r2_score(y, y_pred_manual)

print(f"\nManual MSE: {mse_manual:.6f}")
print(f"Manual R^2: {r2_manual:.6f}")

# Plot Actual vs Predicted (manual)
plt.figure(figsize=(6,5))
plt.scatter(y, y_pred_manual)
plt.xlabel('Actual Final Exam Score')
plt.ylabel('Predicted (manual)')
plt.title('Actual vs Predicted (Manual)')
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'k--', lw=1)
plt.show()
```

## 4. scikit-learn implementation (for comparison)

```python
model = LinearRegression()
model.fit(X, y)

y_pred_sklearn = model.predict(X)

print('\nScikit-learn coefficients:')
print('  intercept:', model.intercept_)
for i, coef in enumerate(model.coef_):
    print(f'  coef_{i+1} ({feature_cols[i]}): {coef:.6f}')

mse_sklearn = mean_squared_error(y, y_pred_sklearn)
r2_sklearn = r2_score(y, y_pred_sklearn)

print(f"\nSklearn MSE: {mse_sklearn:.6f}")
print(f"Sklearn R^2: {r2_sklearn:.6f}")

# They should be identical or extremely close to manual results
```

## 5. 3D Visualization for two features at a time
We'll plot Hours_Studied (x), Sleep_Hours (y) vs Final_Exam_Score (z) and overlay predicted surface.

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(9,7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['Hours_Studied'], df['Sleep_Hours'], df[target_col], c='blue', label='Actual')

# Create a grid for the surface
x_surf = np.linspace(df['Hours_Studied'].min(), df['Hours_Studied'].max(), 20)
y_surf = np.linspace(df['Sleep_Hours'].min(), df['Sleep_Hours'].max(), 20)
X_surf, Y_surf = np.meshgrid(x_surf, y_surf)

# For surface Z = theta0 + theta1*x + theta2*y + theta3*(other features average)
# We need to fix Past_Score at its mean for surface plotting
past_mean = df['Past_Score'].mean()

Z_surf = (theta_manual[0] + theta_manual[1]*X_surf + theta_manual[2]*Y_surf + theta_manual[3]*past_mean)

ax.plot_surface(X_surf, Y_surf, Z_surf, alpha=0.4, cmap='viridis')
ax.set_xlabel('Hours_Studied')
ax.set_ylabel('Sleep_Hours')
ax.set_zlabel('Final Exam Score')
ax.set_title('Regression Plane (Hours vs Sleep)')
plt.show()
```

## 6. Interactive predictor using ipywidgets
 Create sliders based on min/max of each feature for comfortable ranges

```python
sliders = {}
for col in feature_cols:
    col_min, col_max = float(df[col].min()), float(df[col].max())
    span = col_max - col_min
    # add some margin
    sliders[col] = widgets.FloatSlider(value=(col_min+col_max)/2,
                                       min=max(0, col_min - 0.2*span),
                                       max=col_max + 0.2*span,
                                       step=0.1,
                                       description=col)

out = widgets.Output()

def update_prediction(change=None):
    with out:
        out.clear_output()
        vals = [1.0] + [sliders[c].value for c in feature_cols]
        vals = np.array(vals)
        pred_manual = vals.dot(theta_manual)
        pred_sklearn = vals[1:].dot(model.coef_) + model.intercept_
        display(Markdown(f"**Manual prediction:** {pred_manual:.2f}  **Sklearn prediction:** {pred_sklearn:.2f}"))

# Attach observers
for s in sliders.values():
    s.observe(update_prediction, names='value')

# Display widgets and output
for s in sliders.values():
    display(s)

display(out)

# Initial update
update_prediction()
```

## 7. Save coefficients and provide a small predict function (useful for future cells)

```python
def predict_manual_dict(feature_dict):
    # feature_dict: {feature_name: value}
    vals = [1.0] + [feature_dict[c] for c in feature_cols]
    return np.array(vals).dot(theta_manual)

# Save coefficients to a small CSV for documentation
coef_df = pd.DataFrame({
    'feature': ['intercept'] + feature_cols,
    'coefficient': theta_manual
})
coef_df.to_csv('manual_coefficients.csv', index=False)

print('\nSaved manual coefficients to manual_coefficients.csv')
```
